```python
def getM(s):
  return float(s.replace(',','.').split()[0])
```

```python
def getMW(s):
  try:
    w = float(s.replace(',','.').split()[0])
    m = round(w/4,2)
  except:
    m = 0
  return m
```

```python
def getMD(s):
  try:
    d= float(s.replace(',','.').split()[0])
    m = round(d/30,2)
  except:
    m = 0
  return m
```

```python
def getMH(s):
  try:
    h= float(s.replace(',','.').split()[0])
    m = round(h/64,2)
  except:
    m = 0
  return m
```

```python
def getMY(s):
  try:
   y= float(s.replace(',','.').split()[0])
   m = round(y*12,2)
  except:
    m = 0
  return m
```

```python
def getMU(s):
  try:
   y= float(s.replace(',','.').split()[0])
   m = round(y/32,2)
  except:
    m = 0
  return m
```
